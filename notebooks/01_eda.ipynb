# Conclusiones finales del EDA · DataCo
 
- **Calidad de Datos y Gestión de Nulos:** La integridad general del dataset es excelente en la mayoría de sus variables transaccionales y financieras. La excepción principal se concentra en la columna `Order Zipcode`, la cual presenta valores nulos que requieren una estrategia de imputación o limpieza en la fase ETL, dado que pueden afectar el análisis geoespacial del destino de las órdenes.
 
- **Duplicados y Estructura Lógica:** Las repeticiones detectadas en los registros no constituyen errores de duplicidad aleatoria; se justifican plenamente por la estructura lógica y granular del dataset, donde un mismo número de orden (`Order Id`) puede contener múltiples ítems de productos individuales o registrar distintas actualizaciones en el historial de envío.
 
- **Políticas de Privacidad y Reducción de Dimensiones (Eliminación):** Con el fin de optimizar el rendimiento del almacenamiento en el ecosistema de Big Data y garantizar el cumplimiento estricto de las normativas de protección de datos (debido a la naturaleza sintética/expuesta de cierta información), se determina la eliminación definitiva de las siguientes columnas no relevantes o sensibles:
  - `Customer Email` (Privacidad)
  - `Customer Password` (Seguridad)
  - `Product Description` (Columna redundante o vacía en el catálogo de productos)
  - `Order State` (Duplicidad con las métricas regionales o de envío)
 
- **Estandarización y Corrección de Anomalías:** Se detectaron inconsistencias críticas en la columna de ubicación del cliente (`Customer State`). Mientras que la mayoría de los registros adoptan el formato estándar de código de estado de dos letras (ej. NY, TX), se identificaron códigos postales numéricos (como *95758* y *91732*) en dicho campo. Como acción obligatoria de limpieza, se plantea mapear y reemplazar estos números por el estado correcto al que pertenecen (`CA`).
 
- **Optimización de Estructura y Formato de Columnas:** Dado que las columnas ya se encuentran en el idioma inglés, la transformación se centrará en estandarizar su formato de texto. Se aplicará una conversión estricta a minúsculas y se reemplazarán los espacios por guiones bajos (formato `snake_case`) para homogeneizar los nombres de las variables (ej. de `Order Id` a `order_id`). Asimismo, se transformarán las variables temporales desde texto plano a formato `datetime` nativo.
 
- **Integración y Complementariedad:** Las tablas analizadas demuestran ser altamente complementarias para construir un modelo en estrella o copo de nieve, permitiendo cruzar de manera directa las dimensiones de Clientes, Productos y Geografía con la tabla de hechos de Órdenes y Envíos.
