```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
import sys
import os # Esto sube un nivel en las carpetas para encontrar la raíz del proyecto
sys.path.append(os.path.abspath(os.path.join('..')))
import src.exploration as e
import src.transform as t
```

```python
warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

```python
df_data = pd.read_csv("../files/raw/dataco_supply_chain.csv", sep=None, engine="python", encoding="latin-1")
```

```python
df_data = t.standardize_columns(df_data)
```

```python
df_data.sample(5)
```

```python
# Convertimos la columna al formato datetime correcto
df_data['order_date_dateorders'] = pd.to_datetime(df_data['order_date_dateorders'])
```

```python
# Esto te devolverá ambas fechas en una sola salida limpia

df_data['order_date_dateorders'].agg(['min', 'max'])
 
```

#### Eliminar columnas

```python
df_data.drop(columns=["product_status", "customer_email", "customer_password", "product_description", "order_zipcode", "market", "order_state","product_image"],inplace=True)
```

#### Cambiar los códigos por abreviaturas del estado

```python
dicc_ca = {"95758" : "CA", "91732" : "CA"}

df_data["customer_state"] = df_data["customer_state"].map(dicc_ca).fillna(df_data["customer_state"])
```

```python
df_data["customer_state"].unique()
```

#### Convertir tipos de datos

```python
df_data["customer_zipcode"] = df_data["customer_zipcode"].astype(str).str.replace(".0", "", regex=False)
```

```python
df_data["order_item_discount"] = df_data["order_item_discount"].astype(float)
```

#### Rellenar los 3 nulos de la columna customer_zipcode usando la latitud y la longitud

```python
# 1

df_data.loc[np.isclose(df_data["latitude"], 39.134159) & np.isclose(df_data["longitude"], -123.718552),"customer_zipcode"] = "95410"
 
# 2

df_data.loc[np.isclose(df_data["latitude"], 38.408573) & np.isclose(df_data["longitude"], -121.449112),"customer_zipcode"] = "95823"
 
# 3

df_data.loc[np.isclose(df_data["latitude"], 34.071892) & np.isclose(df_data["longitude"], -118.013886),"customer_zipcode"] = "91732"
 
```

#### Aplicar función para redondear los decimales

```python
t.round_up(df_data)

df_data.head()
```

#### Dividir una columna en dos, fecha y hora y cambiar el formato

```python
df_data[['date', 'datetime']] = df_data['shipping_date_dateorders'].str.split(' ', expand=True)
```

```python
df_data.head()
```

```python
df_data.drop(columns=["shipping_date_dateorders"],inplace=True)
```

```python
df_data.head()
```

```python
# Convertimos la columna al formato datetime correcto
df_data["date"] = pd.to_datetime(df_data["date"]).dt.strftime("%m-%d-%Y")
df_data["datetime"] = pd.to_timedelta(df_data["datetime"] + ":00").astype(str).str[-8:-3]
```

```python
df_data.head()
```

#### Rellenar los nulos de la columna customer_lname por unknown

```python
df_data["customer_lname"] = df_data["customer_lname"].fillna("Unknown")
```

```python
df_data["customer_lname"].isna().sum()
```

```python
df_data.to_csv("../files/processed/dataco_supply_chain.csv", index=False)
```
